Both China and Russia's relationships with Myanmar have implications for the United States. Myanmar's strategic location makes it an important player in regional geopolitics, and influence over the country allows China and Russia to extend their reach. For the U.S., it's essential to monitor these relationships to understand the geopolitical dynamics in the region and strategically respond to any potential challenges or opportunities. ## **Issues & Future Improvements**

There are a few issues came across during the analysis process.


*  An identified issue pertains to language variations. Running analyses across all languages, except for the last graph, has brought attention to misspellings and potential variations in word generation. Notably, different spellings of the same words may go unnoticed as key words. This observation underscores the need for a more refined approach in separating language sources during subsequent analyses to enhance accuracy and granularity.


*   An additional issue arises from the duplication of information, where some news articles extract and expand on a piece of information from another, leading to potential inaccuracies in the percentage calculations. To address this challenge, human validation becomes crucial, ensuring the accuracy and reliability of the data by cross-referencing and verifying the information extracted from different sources.
